```python
import pandas as pd

# -------- CONFIG --------
expa_role = str(row.get("opportunity title", "")).strip()
print("Role Title:", expa_role)
# ------------------------

# Load CSV
df = pd.read_csv(csv_path)
df.columns = df.columns.str.strip().str.lower()

row = df.iloc[SHEET_ROW_NUMBER - 2]
title = row["opportunity title"].lower()

def classify_role(title):
    # ⬇️ PASTE INVALID CHECK HERE
    title = title.lower().strip()

    invalid_exact_titles = {...}

    if title in invalid_exact_titles:
        return "❌ INVALID TITLE (Needs Manual Fix)"


# PHARMACEUTICAL / CLINICAL / DRUG DOMAIN
    if any(k in title for k in [
        "pharma", "pharmacy", "drug", "clinical",
        "regulation", "regulatory", "cocrystal",
        "anticancer", "in vitro", "screening",
        "nash", "formulation"
    ]):
        return "Pharmaceutical Intern"

    # -------- BIOMEDICAL / BIO --------
    if any(k in title for k in [
        "biomedical", "biosensing", "respiratory", "artery",
        "stent", "patient", "clinical", "wound",
        "cardio", "health monitoring", "mr image", "zebrafish"
    ]):
        return "Biomedical Engineering Intern"

    # -------- BIO / LIFE SCIENCES --------
    if any(k in title for k in [
        "bioinformatics", "in vitro", "molecular",
        "medicinal plant", "secondary metabolite", "toxicity"
    ]):
        return "Biological Sciences Intern"

    # -------- MECHANICAL / CFD / FEA --------
    if any(k in title for k in [
        "cfd", "finite element", "fem", "fea",
        "aerodynamics", "thermal", "structures",
        "pneumatic", "sandwich structure"
    ]):
        return "Mechanical Engineering Intern"

    # -------- ELECTRICAL / ELECTRONICS --------
    if any(k in title for k in [
        "fpga", "microcontroller", "pwm", "motor",
        "smart grid", "control system", "embedded",
        "cyber security", "quantum computing"
    ]):
        return "Electrical & Electronics Engineering Intern"

    # -------- MATERIALS / CHEM / PHARMA --------
    if any(k in title for k in [
        "nanomaterial", "thin film", "coating",
        "catalyst", "co-crystal", "anticancer",
        "pharmaceutical", "drug", "screening"
    ]):
        return "Chemical / Materials Science Intern"

    # -------- ENERGY / EV / ENVIRONMENT --------
    if any(k in title for k in [
        "energy", "electric vehicle", "wind",
        "renewable", "storage", "e-mobility",
        "carbon dioxide", "emission", "green cover"
    ]):
        return "Energy / Environmental Engineering Intern"

    # -------- DATA / AI / COMPUTATION --------
    if any(k in title for k in [
        "computational", "deep learning", "machine learning",
        "optimization", "multi-agent", "tracking",
        "digital twin", "simulation", "model"
    ]):
        return "Data / AI Engineering Intern"

    # -------- ARCHITECTURE / DESIGN --------
    if any(k in title for k in [
        "architecture", "architectural", "design"
    ]):
        return "Architecture / Design Intern"

    # -------- SOCIAL SCIENCE / POLICY --------
    if any(k in title for k in [
        "education", "disabilities", "religion",
        "women", "pedestrian", "cities", "policy",
        "transcultural"
    ]):
        return "Social Sciences / Policy Intern"

    # -------- BUSINESS / FINANCE --------
    if any(k in title for k in [
        "banking", "finance", "consumer",
        "marketing", "management", "business"
    ]):
        return "Business / Finance Intern"

    # -------- FALLBACK --------
    return "General Engineering Intern"


final_title = classify_role(title)

print("ORIGINAL TITLE:")
print(row["opportunity title"])

print("\nEXPA ROLE TITLE:")
print(final_title)
```
